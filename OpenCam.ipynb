program ini menggunakan OpenCV untuk mendeteksi wajah dalam gambar yang diambil dari webcam dan model CNN yang telah dilatih untuk mengklasifikasikan emosi berdasarkan ekspresi wajah. Pertama, file Haar Cascade yang berfungsi untuk deteksi wajah dimuat menggunakan cv2.CascadeClassifier. Kemudian, model yang telah dilatih disimpan dalam format .h5 dimuat menggunakan load_model dari Keras. Label emosi, seperti 'Angry', 'Disgust', 'Fear', dan lain-lain, disiapkan untuk memetakan hasil prediksi. Proses deteksi wajah dilakukan dalam loop video, di mana setiap frame dari webcam diubah menjadi citra grayscale. Wajah dalam citra tersebut dideteksi menggunakan detectMultiScale.

Setelah wajah terdeteksi, gambar wajah tersebut (region of interest atau ROI) diproses dengan melakukan perubahan ukuran ke 48x48 piksel dan normalisasi. Citra wajah tersebut kemudian dikonversi menjadi array menggunakan img_to_array, dan dimasukkan ke dalam model CNN untuk memprediksi emosi. Hasil prediksi dikonversi ke label emosi menggunakan emotion_labels dan ditampilkan pada gambar. Jika tidak ada wajah yang terdeteksi, maka teks 'No Face Found' akan ditampilkan di layar. Tampilan hasil prediksi (label emosi) ditambahkan di atas wajah pada frame dengan cv2.putText. Program ini terus berjalan dalam loop hingga pengguna menekan tombol 'q' untuk keluar dari aplikasi. Kamera dihentikan dan semua jendela OpenCV ditutup dengan cap.release() dan cv2.destroyAllWindows().

```python
from keras.models import load_model
from time import sleep
from keras.preprocessing.image import img_to_array
from keras.preprocessing import image
import cv2
import numpy as np

# Load Haar Cascade dan model
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') # file haarcascade_frontalface_default.xml didownload dari github opencv
classifier = load_model('model.h5')

# Label emosi
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

# Buka kamera
cap = cv2.VideoCapture(0)

# Periksa apakah kamera terbuka

while True:
    _, frame = cap.read()
    labels = []

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_gray = cv2.resize(roi_gray, (48, 48), interpolation=cv2.INTER_AREA)

        if roi_gray is not None:
            roi = roi_gray.astype('float') / 255.0
            roi = cv2.resize(roi, (48,48))
            roi = img_to_array(roi)
            roi = np.expand_dims(roi, axis=0)

            prediction = classifier.predict(roi)[0]
            print(f"Prediction: {prediction}")
            print(f"Label: {emotion_labels[prediction.argmax()]}")
            label = emotion_labels[prediction.argmax()]
            label_position = (x, y-10)
            cv2.putText(frame, label, label_position, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        else:
            cv2.putText(frame, 'No Face Found', (30, 80), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow('Emotion Detector', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'): #tekan q untuk menutup kamera
        break

cap.release()
cv2.destroyAllWindows()
```
